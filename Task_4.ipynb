```python
import pandas as pd
dataset = pd.read_csv('/content/sentiment_data.csv')
dataset.head()
```

```python
import pandas as pd
import csv
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
dataset
```

```python
tokenizer = Tokenizer(num_words = 5000)
tokenizer.fit_on_texts(dataset['text'])

train_sequences = tokenizer.texts_to_sequences(dataset['text'])
train_padded = pad_sequences(train_sequences, maxlen=200)
```

```python
train_text, val_text, train_labels, val_labels = train_test_split(train_padded, dataset['label'], test_size=0.2, random_state=42)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Embedding(input_dim=5000, output_dim=128, input_length=200),
    tf.keras.layers.LSTM(64, dropout=0.2),
    tf.keras.layers.Dense(3, activation='softmax')
])
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
import numpy as np

# Assuming train_labels and val_labels are pandas Series
train_labels = np.asarray(train_labels).astype('object')
val_labels = np.asarray(val_labels).astype('object')


model.fit(train_text, train_labels, epochs=2, batch_size= 32, validation_data=(val_text, val_labels))
```
